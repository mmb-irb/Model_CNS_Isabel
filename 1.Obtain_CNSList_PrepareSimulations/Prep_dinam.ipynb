# 1. Write to each PDB folder the .sh files needed for running the dynamic

```python
# We write the file line by line that we want to add to the directories
# of each PDB. 
import os 
```

```python
PDBs_name = []
for file in os.listdir('/home/imartinv/Escritorio/TFG/1.LIST_PROTEINS/Isa_List_proteins/AllJupyters/2.Filters/2.5.MemprotMD/Memprot_DruggablePD/ToStudy_NonRepeatedSequences/unzipped'):
    PDB_n = file
    PDBs_name.append(PDB_n)
```

```python
# File 1: sbatch_P1_script.sh

line1_a = "#!/bin/bash"
line2_a = '#SBATCH --time=72:00:00'
line3_a = '#SBATCH --error=error1.log'
line4_a = '#SBATCH --output=output.log'
line5_a = '#SBATCH --ntasks=1'
line6_a = '#SBATCH --cpus-per-task=40'
line7_a = '#SBATCH --ntasks-per-socket=1'
line8_a = 'module load fftw/3.3.8'
line9_a = 'module load gromacs/2022.3'
line10_a = 'mkdir EM'
line11_a = 'gmx_mpi grompp -f mdp_files/em.mdp -o EM/em -c atomistic-system.pdb -n atomistic-system.ndx -maxwarn -1'
line12_a = 'cd EM'
line13_a = 'gmx_mpi mdrun -deffnm em -v'
```

```python
# File 2: sbatch_P2_script.sh

line1_b = "#!/bin/bash"
line2_b = '#SBATCH --time=72:00:00'
line3_b = '#SBATCH --error=error2.log'
line4_b = '#SBATCH --output=output.log'
line5_b = '#SBATCH --ntasks=1'
line6_b = '#SBATCH --cpus-per-task=40'
line7_b = '#SBATCH --ntasks-per-socket=1'
line8_b = 'module load fftw/3.3.8'
line9_b = 'module load gromacs/2022.3'
line10_b = 'mkdir PR'
line11_b = 'gmx_mpi grompp -f mdp_files/pr.mdp -o PR/pr -c EM/em.gro -r EM/em.gro -n atomistic-system.ndx -maxwarn -1'
line12_b = 'cd PR'
line13_b = 'gmx_mpi mdrun -deffnm pr -v'
```

```python
# File 3: sbatch_P3_script.sh

line1_c = "#!/bin/bash"
line2_c = '#SBATCH --time=72:00:00'
line3_c=  '#SBATCH --error=error3.log'
line4_c = '#SBATCH --output=output.log'
line5_c = '#SBATCH --ntasks=1'
line6_c = '#SBATCH --cpus-per-task=40'
line7_c = '#SBATCH --ntasks-per-socket=1'
line8_c = 'module load fftw/3.3.8'
line9_c = 'module load gromacs/2022.3'
line10_c = 'mkdir 100ns'
line11_c = 'gmx_mpi grompp -f mdp_files/100ns.mdp -o 100ns/md -c PR/pr.gro -n atomistic-system.ndx -maxwarn -1'
line12_c = 'cd 100ns'
line13_c = 'gmx_mpi mdrun -deffnm md -v'
```

```python
# File 4: sbatch_cont_script.sh

line1_d = "#!/bin/bash"
line2_d = '#SBATCH --time=72:00:00'
line3_d=  '#SBATCH --error=error3.log'
line4_d = '#SBATCH --output=output.log'
line5_d = '#SBATCH --ntasks=1'
line6_d = '#SBATCH --cpus-per-task=40'
line7_d = '#SBATCH --ntasks-per-socket=1'
line8_d = 'module load fftw/3.3.8'
line9_d = 'module load gromacs/2022.3'
line10_d = 'cd 100ns'
line11_d = 'gmx_mpi mdrun -deffnm md -v -cpi md.cpt'
```

```python

for PDB_name_file in PDBs_name:
    
    
    common_path = '/home/imartinv/Escritorio/TFG/1.LIST_PROTEINS/Isa_List_proteins/AllJupyters/2.Filters/2.5.MemprotMD/Memprot_DruggablePD/ToStudy_NonRepeatedSequences/unzipped/'

    save_path = common_path+PDB_name_file
    
    PDB_name = PDB_name_file[0:4]
    
    if PDB_name != "2k58" and PDB_name != "2ksr" and PDB_name != "2lm2" and PDB_name != "2r4s" and PDB_name != "6j8e":
    

        #File 1

        file_name = "sbatch_P1_script_"+PDB_name+".sh"

        completeName = os.path.join(save_path, file_name)

        file1 = open(completeName, "w")

        file1.write("%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n" % (line1_a, line2_a, line3_a, line4_a,line5_a,line6_a,line7_a,line8_a,line9_a,line10_a,line11_a,line12_a,line13_a))

        file1.close()

        #File 2

        file_name = "sbatch_P2_script_"+PDB_name+".sh"

        completeName = os.path.join(save_path, file_name)

        file2 = open(completeName, "w")

        file2.write("%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n" % (line1_b, line2_b, line3_b, line4_b,line5_b,line6_b,line7_b,line8_b,line9_b,line10_b,line11_b,line12_b,line13_b))

        file2.close()

        #File 3

        file_name = "sbatch_P3_script_"+PDB_name+".sh"

        completeName = os.path.join(save_path, file_name)

        file3 = open(completeName, "w")

        file3.write("%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s" % (line1_c, line2_c, line3_c, line4_c,line5_c,line6_c,line7_c,line8_c,line9_c,line10_c,line11_c,line12_c,line13_c))

        file3.close()


        #File 4. continue

        file_name = "sbatch_cont_script_"+PDB_name+".sh"

        completeName = os.path.join(save_path, file_name)

        file4 = open(completeName, "w")

        file4.write("%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n%s\n" % (line1_d, line2_d, line3_d, line4_d,line5_d,line6_d,line7_d,line8_d,line9_d,line10_d,line11_d))

        file4.close()

```
